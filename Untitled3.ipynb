```python
#Capping outliers

#code attribution


#this code was taken from youtube

#https://eur03.safelinks.protection.outlook.com/?url=https%3A%2F%2Fyoutube.com%2Fshorts%2FmO4VaM5niq0%3Fsi%3DZIJTxhyr5Dl8MEPE&data=05%7C02%7CST10480399%40vcconnect.edu.za%7C35b339d868dd4f61467a08dd8403238c%7Ce10c8f44f469448fbc0dd781288ff01b%7C0%7C0%7C638811871233163698%7CUnknown%7CTWFpbGZsb3d8eyJFbXB0eU1hcGkiOnRydWUsIlYiOiIwLjAuMDAwMCIsIlAiOiJXaW4zMiIsIkFOIjoiTWFpbCIsIldUIjoyfQ%3D%3D%7C0%7C%7C%7C&sdata=5Ox9hc1LBg1ap%2FYQH1mtQJK0Tv8VV1d9jwT1rJJEn9I%3D&reserved=0

#Caps outliers in a DataFrame column using the IQR method.
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[column] = np.where[(df[column] > upper,upper,np.where(df[column] < lower,lower,df[column]))]
    return df
```
